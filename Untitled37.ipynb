```python
import cv2
from google.colab.patches import cv2_imshow  # For displaying images in Google Colab

# Configuration
frame_path = '/content/hotelimage-43/test/images/22_DDNS1_jpg.rf.d55db328c901b8e0fd830d3c3dac54d6.jpg'  # Change to the path of your image
confidence_threshold = 0.5 # Confidence threshold for predictions

# Update this dictionary to change the class names as desired
class_names = {0: 'bed', 1: 'chair', 2: 'couch', 3: 'sofa', 4: 'table', 5: 'teapoy'}

def predict_with_confidence(model, frame_path, confidence_threshold):
    # Load the frame
    frame = cv2.imread(frame_path)
    if frame is None:
        print("Error: Frame not found or path is incorrect.")
        return

    # Run prediction using the specified confidence threshold
    results = model(frame, conf=confidence_threshold)  # Assuming your model takes `conf` as an argument

    # Process predictions
    for i, result in enumerate(results):
        print(f"Processing result {i+1}/{len(results)}...")  # Debug info

        for j, box in enumerate(result.boxes):
            conf = float(box.conf.cpu().numpy().item())
            if conf >= confidence_threshold:  # Ensure predictions meet the threshold
                class_id = int(box.cls.cpu().numpy().item())
                coords = box.xyxy.cpu().numpy().astype(int).flatten()

                if coords.size == 4:
                    x1, y1, x2, y2 = coords

                    # Draw bounding box and label
                    class_name = class_names.get(class_id, f'Class {class_id}')  # Retrieve custom class name
                    label = f'{class_name} (ID: {class_id}), Confidence: {conf:.2f}'

                    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                    cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

                    # Print detection details
                    print(f"Object {j+1}:")
                    print(f"Class ID: {class_id}")
                    print(f"Confidence: {conf:.2f}")
                    print(f"Bounding Box: ({x1}, {y1}, {x2}, {y2})")
                    print("-" * 30)

    # Display the output image with detected boxes
    cv2_imshow(frame)


# Replace 'model' with your actual model instance
predict_with_confidence(model, frame_path, confidence_threshold)
```
